```python
import pickle
with open('log_reg_model.pkl', 'rb') as f:
  log_reg_new = pickle.load(f)
```

```python
import pickle
with open('scaler.pkl', 'rb') as f:
  scaler_new = pickle.load(f)
```

```python
import pandas as pd

```

```python
new_df = pd.read_csv("https://raw.githubusercontent.com/Geetanjali-Kumari1/FutureIQ/main/new_customers.csv")
```

```python
new_df
```

```python
len(new_df)
```

```python
one_hot_encoded_df_new = pd.get_dummies(new_df, columns=['Gender'])
```

```python
one_hot_encoded_df_new['Gender_Female'] = one_hot_encoded_df_new['Gender_Female'].astype(int)
one_hot_encoded_df_new['Gender_Male'] = one_hot_encoded_df_new['Gender_Male'].astype(int)
```

```python
one_hot_encoded_df_new
```

```python
one_hot_encoded_df_new = one_hot_encoded_df_new.drop(columns=['Gender_Female'])
```

```python
one_hot_encoded_df_new
```

```python
print(one_hot_encoded_df_new.head())
```

```python
df_with_gender_new = pd.concat([new_df['Gender'], one_hot_encoded_df_new], axis=1)
```

```python
print(df_with_gender_new.head())
```

```python
x_new = df_with_gender_new[['Gender_Male','Age','Salary','Price']].to_numpy()
```

```python
x_new
```

```python
x_new_scale2 = scaler_new.fit_transform(x_new)
```

```python
x_new_scale2
```

```python
y_new_pred = log_reg_new.predict(x_new_scale2)
```

```python
y_new_pred
```

```python
df_with_gender_new['will_purchase'] = y_new_pred
```

```python
df_with_gender_new
```

```python
df_with_gender_new.to_csv("log_reg_model_predictions.csv",index=False)
```
